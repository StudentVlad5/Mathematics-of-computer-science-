Function 1

$f_1(x) = \frac{x^3}{3} + \frac{x^2}{2} + 2x$

Знайдемо похідну кожного доданка:
$\frac{d}{dx}(\frac{x^3}{3}) = x^2\\
\frac{d}{dx}(\frac{x^2}{2}) = x\\
\frac{d}{dx}(2x) = -2$

$f'_1(x) = x^2 + x - 2$

$f'_1(1) = 1^2 + 1 - 2 = 0

$f'_1(-\frac{1}{2}) = (-\frac{1}{2})^2 + -\frac{1}{2} - 2 = -2.25$

Function 2

$f_2(x) = \sqrt{x^2 + 1}$

Використаємо правило похідної степеневої функції:
$f'_2(x) = (x^2 + 1)^\frac{1}{2} = \frac{1}{2}(x^2 + 1)^\frac{-1}{2} * 2x = \frac{x}{x^2 + 1}$
$f'_2(1) = (\frac{1}{\sqrt{2}}) = 0.7071$
$f'_2(-\frac{1}{2}) = (\frac{-\frac{1}{2}}{\sqrt{\frac{1}{4} + 1}}) = -0.4472$

Function 3

$f_3(x) = \frac{1}{\sqrt{x^2 + 1}}$

Використаємо правило похідної степеневої функції:
$f'_3(x) = (x^2 + 1)^\frac{-1}{2} = -\frac{1}{2}(x^2 + 1)^\frac{-3}{2} * 2x = -\frac{x}{(x^2 + 1)^\frac{3}{2}}$
$f'_3(1) = -\frac{1}{(1^2 + 1)^\frac{3}{2}} = -0.3536$
$f'_3(-\frac{1}{2}) = -\frac{-\frac{1}{2}}{(-\frac{1}{2}^2 + 1)^\frac{3}{2}} = 0.3578$


```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Крок 1: Символьна змінна
x = sp.symbols('x')

# Крок 2: Задання функцій
f1 = x**3 / 3 + x**2 / 2 - 2*x
f2 = sp.sqrt(x**2 + 1)
f3 = 1 / sp.sqrt(x**2 + 1)

# Крок 3: Похідні функцій
f1_prime = sp.diff(f1, x)
f2_prime = sp.diff(f2, x)
f3_prime = sp.diff(f3, x)

# Крок 4: Значення похідних у точках x = 1 і x = -1/2
points = [1, -1/2]
print("Значення похідних у точках:")
for pt in points:
    print(f"x = {pt}")
    print(f"  f1'({pt}) =", f1_prime.subs(x, pt).evalf())
    print(f"  f2'({pt}) =", f2_prime.subs(x, pt).evalf())
    print(f"  f3'({pt}) =", f3_prime.subs(x, pt).evalf())
    print()

# Крок 5: Побудова графіків
x_vals = np.linspace(-3, 3, 400)
f1_func = sp.lambdify(x, f1, "numpy")
f1p_func = sp.lambdify(x, f1_prime, "numpy")
f2_func = sp.lambdify(x, f2, "numpy")
f2p_func = sp.lambdify(x, f2_prime, "numpy")
f3_func = sp.lambdify(x, f3, "numpy")
f3p_func = sp.lambdify(x, f3_prime, "numpy")

# Побудова підграфіків
fig, axes = plt.subplots(3, 2, figsize=(12, 12))

# Графіки функцій
axes[0, 0].plot(x_vals, f1_func(x_vals), label="f1(x)", color='blue')
axes[0, 0].set_title("f1(x) = x^3/3 + x^2/2 - 2x")
axes[0, 0].grid()

axes[0, 1].plot(x_vals, f1p_func(x_vals), label="f1'(x)", color='red')
axes[0, 1].set_title("f1'(x)")
axes[0, 1].grid()

axes[1, 0].plot(x_vals, f2_func(x_vals), label="f2(x)", color='blue')
axes[1, 0].set_title("f2(x) = sqrt(x^2 + 1)")
axes[1, 0].grid()

axes[1, 1].plot(x_vals, f2p_func(x_vals), label="f2'(x)", color='red')
axes[1, 1].set_title("f2'(x)")
axes[1, 1].grid()

axes[2, 0].plot(x_vals, f3_func(x_vals), label="f3(x)", color='blue')
axes[2, 0].set_title("f3(x) = 1 / sqrt(x^2 + 1)")
axes[2, 0].grid()

axes[2, 1].plot(x_vals, f3p_func(x_vals), label="f3'(x)", color='red')
axes[2, 1].set_title("f3'(x)")
axes[2, 1].grid()

plt.tight_layout()
plt.show()
```

$y=-3x^2 + 30x\\
\frac{dy}{dx} = \frac{d}{dx}(-3x^2 + 30x)\\
\frac{d}{dx}(-3x^2) = -3 * 2 * x^{2-1} = -6x\\
\frac{d}{dx}(30x) = 30\\
\frac{dy}{dx} = -6x + 30\\
$

Знайдемо  точку  максимуму

$ -6x + 30 = 0 \\
x = 5$

Перевіримо характер точки (максимум чи мінімум)

$\frac{d^2y}{dx^2} = \frac{d}{dx}(-6x + 30) = -6\\
$

Оскільки друга похідна від'ємна, це підтверджує, що в точці 𝑥 = 5  функція має максимум.

```python
import numpy as np
import matplotlib.pyplot as plt

# Визначаємо функцію та її похідну
def f(x):
    return -3*x**2 + 30*x

def df(x):
    return -6*x + 30

# Створюємо значення x для графіка
x_vals = np.linspace(0, 10, 400)
y_vals = f(x_vals)
dy_vals = df(x_vals)

# Точка максимуму
x_max = 5
y_max = f(x_max)

# Створюємо графіки
plt.figure(figsize=(10, 6))

# Графік функції
plt.plot(x_vals, y_vals, label=r'$y = -3x^2 + 30x$', color='blue')

# Графік похідної
plt.plot(x_vals, dy_vals, label=r"$\frac{dy}{dx} = -6x + 30$", color='green', linestyle='--')

# Позначаємо точку максимуму
plt.scatter(x_max, y_max, color='red', zorder=5)
plt.text(x_max + 0.2, y_max, f'Max ({x_max}, {y_max:.2f})', color='red')

# Додаємо підписи
plt.title('Графіки функції та її похідної')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()

# Показуємо графік
plt.show()
```
